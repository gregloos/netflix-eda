```python
#lets make a ratings over time chart
```

```python
cols = ["type", "rating", "date_added"]
tmp  = df[cols].copy()
```

```python
tmp["date_added"] = pd.to_datetime(tmp["date_added"], errors="coerce")
```

```python
tmp["year_added"] = tmp["date_added"].dt.year
```

```python
pivot = (
    tmp
    .groupby(["year_added", "rating"])
    .size()
    .unstack(fill_value=0)
    .sort_index()
)
```

```python
ax = pivot.plot.area(stacked=True, figsize=(10,6), alpha=0.8)
ax.set_title("Netflix catalogue growth by rating")
ax.set_xlabel("Year added")
ax.set_ylabel("Number of titles")
ax.legend(title="Rating", bbox_to_anchor=(1.05,1), loc="upper left")
```
